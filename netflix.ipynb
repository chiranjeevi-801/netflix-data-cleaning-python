### Remove Duplicate Rows 

```python
df=df.drop_duplicates()
df.head()


```

## Handle missing values 

###  Fill missing director 

```python
df['director']=df['director'].fillna('unknown')
df.head()
```

### Fill missing cast 

```python
df['cast']=df['cast'].fillna("Not Available")
df.head()
```

### Fill missing country

```python
df['country']=df['country'].fillna("null")
df.head()
```

### Fill missing rating 

```python
df['rating']=df['rating'].fillna("Not Rated")
```

```python
df.head()
```

### Drop rows where date_added is missing

```python
df=df.dropna(subset=['date_added'])
df.head(10)
```

## Convert Date Column 

```python
df['date_added'] = df['date_added'].str.strip()
df.head()
```

###  Extract year and Month Added

```python
df['date_added'] = df['date_added'].str.strip()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df.head()
```

```python
df.dtypes
```

```python
df['year_added'].value_counts().sort_index()
```

## Clean Duration column 

The duration column contains:

90 min

1 Season

2 Seasons

Split into number and unit.

```python
df[['duration_number','duration_unit']]=df['duration'].str.split(' ',expand=True)
```

```python
# split duration
df[['duration_number','duration_unit']]=df['duration'].str.split(' ',expand=True)

# convert to number
df['duration_number']=pd.to_numeric(df['duration_number'],errors='coerce')

# group analyisis
df.groupby('duration_unit')['duration_number'].mean()
```

```python
df.info()
```

### Clean Genre Column

Genres are separated by commas.

Convert to list:

```python
df['listed_in']=df['listed_in'].str.split(',')
df.head()
```

### Clean Country column

Remove extra workspace

```python
df['country']=df['country'].str.strip()
df.head()
```

### Standardlize Text 
Convert to lowercase (optional but useful).



```python
df['title']=df['title'].str.lower()
df['director']=df['director'].str.lower()
df['country']=df['country'].str.lower()
```

Reset Index

```python
df=df.reset_index(drop=True)
```
